```python
import fitz
from PIL import Image
import os
import io

pdf_path = r"C:\Users\atyaa\OneDrive\Documents\UNESA\ATYA_SAINS DATA SEM 3\DW\analisis-keterkaitan-antara-ipm-dengan-apk-apm-satuan-pendidikan-sd-mi-sederajat-tahun-2024-2025-sd-mi-sederajat.pdf"
output_folder = "images"
os.makedirs(output_folder, exist_ok=True)

doc = fitz.open(pdf_path)

for page_index in range(len(doc)):
    page = doc[page_index]
    image_list = page.get_images(full=True)

    print(f"[+] Halaman {page_index+1}: ditemukan {len(image_list)} gambar")
    if len(image_list) == 0:
        continue
   
    for image_index, img in enumerate(image_list, start=1):
        xref = img[0]  
        base_image = doc.extract_image(xref)
        image_bytes = base_image["image"]
        image_ext = base_image["ext"]
        image = Image.open(io.BytesIO(image_bytes))

        image_filename = f"{output_folder}/img_{page_index+1}_{image_index}.{image_ext}"
        image.save(image_filename, format=image_ext.upper())

        print(f"[*] Gambar disimpan: {image_filename}, ukuran: {image.size[0]}x{image.size[1]}")
```

```python
import pdfplumber
import re

pdf_filename = r"C:\Users\atyaa\OneDrive\Documents\UNESA\ATYA_SAINS DATA SEM 3\DW\analisis-keterkaitan-antara-ipm-dengan-apk-apm-satuan-pendidikan-sd-mi-sederajat-tahun-2024-2025-sd-mi-sederajat.pdf"
with pdfplumber.open(pdf_filename) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        text = page.extract_text() or ""
        urls = re.findall(r'https?://\S+', text)
        for url in urls:
            print(f"Halaman {page_num}: {url}")
```

```python
import pdfplumber
import pandas as pd

pdf_filename = r"C:\Users\atyaa\OneDrive\Documents\UNESA\ATYA_SAINS DATA SEM 3\DW\analisis-keterkaitan-antara-ipm-dengan-apk-apm-satuan-pendidikan-sd-mi-sederajat-tahun-2024-2025-sd-mi-sederajat.pdf"

# List untuk menyimpan semua tabel
all_tables = []

with pdfplumber.open(pdf_filename) as pdf:
    for page_num, page in enumerate(pdf.pages, start=1):
        tables = page.extract_tables()
        if tables:
            print(f"\nHalaman {page_num}: Ditemukan {len(tables)} tabel")
            for table_num, table in enumerate(tables, start=1):
                print(f"\nTabel {table_num} dari halaman {page_num}:")
                # Konversi tabel ke DataFrame
                df = pd.DataFrame(table[1:], columns=table[0])
                print(df)
                # Simpan DataFrame ke list
                all_tables.append({
                    'page': page_num,
                    'table_num': table_num,
                    'data': df
                })
```

```python
# Simpan setiap tabel ke file Excel terpisah
for table_info in all_tables:
    page_num = table_info['page']
    table_num = table_info['table_num']
    df = table_info['data']
    
    # Buat nama file Excel
    excel_filename = f'table_page_{page_num}_table_{table_num}.xlsx'
    
    # Simpan ke Excel
    df.to_excel(excel_filename, index=False)
    print(f"Tabel telah disimpan ke: {excel_filename}")
```
