```python
import pandas as pd

# Membaca file CSV
file_path = 'Jumlah Sekolah, Guru, dan Murid Sekolah Dasar (SD) di Bawah Kementerian Pendidikan, Kebudayaan, Riset,.csv'
df = pd.read_csv(file_path)

# Menampilkan 5 baris pertama
print("Lima baris pertama:")
print(df.head())

# Total jumlah sekolah, guru, dan murid
print("\n Total keseluruhan:")
print("Total Sekolah:", df['Jumlah Sekolah SD (Negeri+Swasta)'].sum())
print("Total Guru:", df['Jumlah Guru SD (Negeri+Swasta)'].sum())
print("Total Murid:", df['Jumlah Murid SD (Negeri+Swasta)'].sum())

# Kabupaten/kota dengan murid terbanyak
murid_terbanyak = df.loc[df['Jumlah Murid SD (Negeri+Swasta)'].idxmax()]
print("\n Kabupaten/Kota dengan jumlah murid terbanyak:")
print(murid_terbanyak[['Kabupaten/Kota', 'Jumlah Murid SD (Negeri+Swasta)']])

# Menambahkan kolom rasio murid per guru
df['Rasio Murid per Guru'] = df['Jumlah Murid SD (Negeri+Swasta)'] / df['Jumlah Guru SD (Negeri+Swasta)']

# Menampilkan 5 kabupaten/kota dengan rasio murid per guru tertinggi
print("\n 5 Kabupaten/Kota dengan rasio murid per guru tertinggi:")
print(df[['Kabupaten/Kota', 'Rasio Murid per Guru']].sort_values(by='Rasio Murid per Guru', ascending=False).head(5))
```
