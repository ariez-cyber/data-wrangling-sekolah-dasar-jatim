```python
import requests
from bs4 import BeautifulSoup

# URL artikel
url = "https://harian.disway.id/read/882782/perbandingan-angka-partisipasi-sekolah-di-jatim-sd-smp-nyaris-sempurna-sma-di-bawah-80-persen"

# Mengirim permintaan HTTP
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

# Parsing HTML
soup = BeautifulSoup(response.text, "html.parser")

# Menemukan elemen judul dan isi artikel
judul = soup.find("h1").text.strip()
isi_artikel = soup.find_all("p")

# Menampilkan hasil
print("Judul:", judul)
print("\nIsi Artikel:")
for paragraf in isi_artikel:
    print(paragraf.text.strip())
```
